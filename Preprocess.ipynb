# Importar librerias

```python
import pandas as pd
import pyarrow.parquet as pq
import csv
import re
import string
from google.colab import drive
drive.mount('/content/drive')
```

# parquet a csv

```python
nombreArchivos = ['/content/drive/My Drive/Colab Notebooks/datos/test.parquet', '/content/drive/My Drive/Colab Notebooks/datos/train.parquet']
nombreCsv = ['/content/drive/My Drive/Colab Notebooks/datosTest.csv', '/content/drive/My Drive/Colab Notebooks/datosTrain.csv']
for i in range(len(nombreArchivos)):
    table = pq.read_table(nombreArchivos[i])
    df = table.to_pandas()
    df.to_csv(nombreCsv[i], index=False)
    with open(nombreCsv[i], newline='') as csvfile:
        reader = csv.reader(csvfile)
        header = next(reader)
        rows = []
        translator = str.maketrans('', '', string.punctuation)
        for row in reader:
            if len(row) > 2:
                first_column = ','.join(row[:-1])
                second_column = row[-1]
                text = first_column.lower()
                text = re.sub(r'\d+', '', text)
                text = text.translate(translator)
                rows.append([text, second_column])
            else:
                rows.append(row)

    df = pd.DataFrame(rows, columns=['review', 'sentiment'])
    df.to_csv(nombreCsv[i], index=False)
```

# Juntar los csv

```python
# Juntar los archivos CSV en uno solo
df1 = pd.read_csv(nombreCsv[0])
df2 = pd.read_csv(nombreCsv[1])

# Combina los dataframes
combined_df = pd.concat([df1, df2], ignore_index=True)

# Guarda el nuevo dataframe combinado en un nuevo archivo CSV
combined_df.to_csv('datos.csv', index=False)
```
